```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv')
data
```

```python
data.info()
```

```python
data = data.drop(['Cabin', 'Ticket', 'PassengerId', 'Name'], axis = 1)
data
```

```python
gender_map={'male':1, 'female': 0}
data['Gender'] = data['Sex'].map(gender_map)
data = data.drop(['Sex'], axis = 1)
data
```

```python
data['Embarked'].fillna(value='Q', inplace=True)
data.info()
```

```python
_map={'Q':1, 'S': 2, 'C' : 3}
data['Embarked_code'] = data['Embarked'].map(_map)
data = data.drop(['Embarked'], axis = 1)
data
```

```python
data1 = data.dropna()
data1.info()
```

```python
data1_na = data.drop(data1.index)
data1_na.info()
```

```python
data1.corr()
```

```python
data1.Age = data1.Age.astype('int')
data1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeRegressor


# Загружаем данные
X = data1.drop(['Age'], axis = 1)
y = data1.Age
# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 483)

    # Обучаем модель
model = DecisionTreeRegressor(max_depth=7)

model.fit(X_train, y_train)

    # Оцениваем модель на тестовой выборке
predictions = model.predict(X_test)
predictions = predictions.astype('int')
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

# Проверяем с какой из таблиц у модели accuracy выше: с таблицей у которой мы сгенерировали недостающие данные или с таблицей где мы убрали пропуски

```python
X = data1.drop(['Survived'], axis = 1)
y = data1.Survived
mx_acc = 0
mx_rs = 0
# Разделяем данные на обучающую и тестовую выборки
for i in range(1, 500):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = i)

    # Обучаем модель
    model = DecisionTreeRegressor(max_depth=5)

    model.fit(X_train, y_train)

    # Оцениваем модель на тестовой выборке
    predictions = model.predict(X_test)
    predictions = predictions.astype('int')
    accuracy = accuracy_score(y_test, predictions)
    if(accuracy > mx_acc):
        mx_acc = accuracy
        mx_rs = i
print(mx_acc)
print(mx_rs)
```
